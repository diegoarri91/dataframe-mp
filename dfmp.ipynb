```python
import numpy as np
import pandas as pd

from dfmp import DataFrameMP
```

```python
def fit(par0=None, par1=None):
    return par0 + par1

class DataFrameFits(DataFrameMP):
    
    def fit(self, idx):
        return fit(**df.loc[idx].to_dict())
```

```python
df = pd.DataFrame(np.random.randn(10, 2), index=['i'+str(ii) for ii in range(10)], columns=['par0', 'par1'])
dfmp = DataFrameFits(df, folder='./data')
display(dfmp.apply_method(dfmp.fit).df)
display(dfmp.df['par0'] + dfmp.df['par1'])
```

```python
import numpy as np
import pandas as pd
import time

def fun1(par0=None, par1=None):
    return par0 + par1

def fun2(par0=None, par1=None, folder=None):
    return [par0 + par1, folder + '2']

def fun3(par0=None, par1=None, folder=None, arg_extra=None):
    return [par0 + par1 + arg_extra, folder + '3']

df = pd.DataFrame(np.random.randn(10, 3), index=['i'+str(ii) for ii in range(10)], columns=['par0', 'par1', 'par2'])
dfmp = DataFrameMP(df, folder='./data')
print(dfmp.apply(fun1, processes=8, columns=['res']).df.head())
print(dfmp.apply(fun2, processes=8, columns=['res', 'string']).df.head())
print(dfmp.apply(fun3, kwargs=dict(arg_extra=100), processes=8, columns=['res', 'string']).df.head())
# print(dfmp.apply(fun2, processes=8, columns=['res']).df.head())
```
